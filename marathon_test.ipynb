```python
from keras.models import Sequential
from keras.layers.core import Dense,Activation
from keras.optimizers import Adam
from sklearn import preprocessing
from keras.models import load_model
model =load_model('dnnformarathon.h5')
```

```python
#feature標準化
#minmax_scale=preprocessing.MinMaxScaler(feature_range=(0,1))
#scaledFeatures=minmax_scale.fit_transform(testdata_fin)
```

```python
#feature標準化
minmax_scale=preprocessing.MinMaxScaler(feature_range=(0,1))
scaledFeatures=minmax_scale.fit_transform(submit_dataset)
```

```python
probability=model.predict(scaledFeatures)
probability=probability.flatten()
probability
```

```python
print(len(probability))
```

```python
pro=[]
for i in range(len(probability)):
    prob=round(probability[i],2)
    pro.append(prob)
```

```python
for item in range(len(pro)):
    print(pro[item])
```

```python
c0=[]
c1=[]
c2=[]
c3=[]
c4=[]
for t in range(len(pro)):
    if t%5==0:
        c0.append(pro[t])
    elif t%5==1:
        c1.append(pro[t])
    elif t%5==2:
        c2.append(pro[t])
    elif t%5==3:
        c3.append(pro[t])
    elif t%5==4:
        c4.append(pro[t])
```

```python
submit_write =pd.read_csv('submit_samples.csv')
submission =pd.DataFrame({
                             "mac_hash":submit_write['mac_hash'],
                             "C0":c0,
                             "C1":c1,
                             "C2":c2,
                             "C3":c3,
                             "C4":c4
                        })
submission.to_csv('submission_marathon.csv',index=False)
```
